### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Heres an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
weather_data = []

# Loop through the scraped data to create a list of rows
for row in rows:
    id = row.find_all('td')[0].text
    terrestrial_date = row.find_all('td')[1].text
    sol = row.find_all('td')[2].text
    ls = row.find_all('td')[3].text
    month = row.find_all('td')[4].text
    min_temp = row.find_all('td')[5].text
    pressure = row.find_all('td')[6].text


    #sending scraped data to the empty array
    weather_data.append({
        'id': id,
        'terrestrial_date': terrestrial_date,
        'sol': sol,
        'ls': ls,
        'month': month,
        'min_temp': min_temp,
        'pressure': pressure
    })    
    
   
#weather_data
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
df = pd.DataFrame(weather_data)
#df     
```

```python
# 1. How many months are there on Mars?
months = len(df['month'].unique())
print(f"There are {months} months in a year on Mars")
```

```python
# 2. How many Martian days' worth of data are there?
data_on_days = len(df['sol'].unique())
print(f"There are {data_on_days} Martian days' worth of data in this dataset.")
```

```python
# 3. What is the average low temperature by month?
avg_min_temp_month = df.groupby('month')['min_temp'].mean().sort_index()
print(f"These are the average low temeperature by month on Mars: {avg_min_temp_month}")
```

```python
# Plot the average temperature by month
avg_min_temp_month.plot(kind='bar')
plt.title('Average Low Temperature by Month on Mars')
plt.xlabel('Month')
plt.ylabel('Temperature (Celsius)')
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location
coldest_month = avg_min_temp_month.idxmin()
hottest_month = avg_min_temp_month.idxmax()

print(f"The coldest month in Curiosity's location is {coldest_month}.")
print(f"The hottest month in Curiosity's location is {hottest_month}.")
```

```python
# 4. Average pressure by Martian month
avg_pressure_by_month = df.groupby('month')['pressure'].mean()
print(avg_pressure_by_month)
```

```python
# Plot the average pressure by month
avg_pressure_by_month.plot(kind='bar')
plt.title('Average Pressure by Month on Mars')
plt.xlabel('Month')
plt.ylabel('Pressure')
plt.show()
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?
totalsols = df['sol'].max() - df['sol'].min() + 1
no_of_years = totalsols / 668.5991 # average sols per Martian year
earth_days_in_martian_year = no_of_years * 687
print("There are roughly {} earth days in a Martian year based on the data in the DataFrame.".format(earth_days_in_martian_year))
```
